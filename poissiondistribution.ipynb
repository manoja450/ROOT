```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial  # Import factorial from SciPy

# Set the mean rate (λ)
lambda_val = 1  

# Generate 10000 random samples from Poisson distribution
data = np.random.poisson(lam=lambda_val, size=10000)

# Plot the histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=range(0, max(data)+1), alpha=0.7, color='blue', edgecolor='black', density=True)

# Overlay the theoretical Poisson PMF
k_values = np.arange(0, max(data)+1)
poisson_pmf = (np.exp(-lambda_val) * lambda_val**k_values) / factorial(k_values)
plt.plot(k_values, poisson_pmf, 'ro-', label='Poisson PMF')

plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution (λ={lambda_val})')
plt.legend()
plt.grid(True)

plt.show()
```
